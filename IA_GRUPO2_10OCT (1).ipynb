```python
def limpiar_texto(texto):
    # Paso 1: Eliminar tildes
    texto_sin_tildes = unidecode(texto)

    # Paso 2: Eliminar caracteres especiales y comas
    texto_limpio = re.sub(r'[^\w\s]', '', texto_sin_tildes)
    texto_limpio = texto_limpio.lower()
    return texto_limpio

def caracteres_especiales(texto):
  texto_limpio = re.sub(r'([a-zA-ZáéíóúÁÉÍÓÚñÑüÜ]+)\d+', r'\1', texto)
  texto_limpio = re.sub(r'\b(?!19|20)\d{1,3}\b', '', texto_limpio)
  texto_limpio = re.sub(r'\s+', ' ', texto_limpio)
  return texto_limpio

def contar_palabras(texto):
    contador = Counter(texto)
    return contador

def palabras_comunes(contador, n=20):
    return contador.most_common(n)

def eliminar_preposiciones(tokens):
    preposiciones = {
        'a', 'ante', 'bajo', 'con', 'contra', 'de', 'desde', 'en', 'entre',
        'hacia', 'hasta', 'para', 'por', 'según', 'sin', 'sobre', 'tras','esta'
    }
    conectores = {
        'y', 'o', 'pero', 'porque', 'aunque', 'sin embargo', 'además', 'así que','la', 'una','de','el','es'
        'un','los','que','del','se','su','las','un','asi'
    }
    # Unir ambas listas
    palabras_a_eliminar = preposiciones.union(conectores)

    return [token for token in tokens if token.lower() not in palabras_a_eliminar]
```
